## How does it connect with past and current work in Multimodal learning?
Multimodal learning models like DeViSE, VirTex, and VisualBERT paved the way for combining image and text understanding. CLIP extended this by using contrastive learning on a massive scale of image-text pairs, enabling robust zero-shot learning.

CLIP also influenced later models such as:

   * Flamingo (DeepMind)
   * BLIP and BLIP-2 (Salesforce)
   * DALL·E and GPT-4V (OpenAI)

## Example
Below is the example of CLIP predicting a caption of the image. <br>
dog.jpeg<br>
It predicts the image's caption and guesses the probability of  the text labels provided being the caption for the image provided.

```python
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import torch

# Load model and processor
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Example image and texts
image = Image.open("dog.jpeg")  # use any local image file
texts = ["a photo of a cat", "a photo of a dog"]

inputs = processor(text=texts, images=image, return_tensors="pt", padding=True)
outputs = model(**inputs)

# Compute similarities
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)
print("Probabilities:", probs)
```

# Results
As we can see ,the probability of the caption being "a photo of a dog" is 0.99 and we can confirm that is true ,as we can see the input image as well ,so without being trained the CLIP model is able to correctly predict the caption of an image.
